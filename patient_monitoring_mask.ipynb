```python
imagem_ref = "referencia.png"
imagem_mov_1 = "movimento_1.png"
imagem_mov_2 = "movimento_2.png"
imagem_mov_3 = "movimento_3.png"
```

```python
lista_arquivos = [
    ("Ref", imagem_ref),
    ("Mov_1", imagem_mov_1),
    ("Mov_2", imagem_mov_2),
    ("Mov_3", imagem_mov_3)
]
```

```python
# Ajustar conforme PixelSpacing do DICOM ou calibração

"""
Transformando "movimento na imagem" em "movimento físico real"
Em imagens médicas DICOM, 1 px = x mm e esse valor vem do equipamento

mm = pixels × PixelSpacing

Se mm é um valor muito alto, pequenos movimentos parecerão enormes
Se mm é um valor muito baixo, grandes movimentos parecerão irrelevantes
"""
PIXEL_TO_MM = 0.1

# Área: (mm²)
PIXEL_TO_MM2 = PIXEL_TO_MM ** 2
```

```python
def carregar_imagem(caminho):
    imagem_bgr = cv2.imread(caminho)

    if imagem_bgr is None:
        raise ValueError(f"Não foi possível carregar a imagem: {caminho}")

    return imagem_bgr
```

```python
def detectar_paciente_mask(imagem_bgr, predictor):
    outputs = predictor(imagem_bgr)
    instances = outputs["instances"].to("cpu")

    if len(instances) == 0:
        return None

    classes = instances.pred_classes.numpy()
    scores = instances.scores.numpy()
    masks = instances.pred_masks

    pessoas = classes == 0
    masks_pessoas = masks[pessoas]
    scores_pessoas = scores[pessoas]

    if len(masks_pessoas) == 0:
        return None

    idx_principal = int(np.argmax(scores_pessoas))

    mask_paciente = masks_pessoas[idx_principal].numpy()
    score_paciente = float(scores_pessoas[idx_principal])

    coords = np.column_stack(np.where(mask_paciente))

    if len(coords) == 0:
        return None

    centroide = coords.mean(axis=0)
    linha_c = int(centroide[0])
    coluna_c = int(centroide[1])

    area = float(mask_paciente.sum())

    return {
        "mask": mask_paciente,
        "score": score_paciente,
        "centroide_linha": float(centroide[0]),
        "centroide_coluna": float(centroide[1]),
        "linha_c": linha_c,
        "coluna_c": coluna_c,
        "area": area
    }
```

```python
def calcular_deslocamento(referencia, atual):
    dx = atual["centroide_coluna"] - referencia["centroide_coluna"]
    dy = atual["centroide_linha"] - referencia["centroide_linha"]
    deslocamento = np.sqrt(dx**2 + dy**2)
    return float(deslocamento)
```

```python
def converter_para_mm(deslocamento_px, area_px):
    deslocamento_mm = deslocamento_px * PIXEL_TO_MM
    area_mm2 = area_px * PIXEL_TO_MM2
    return deslocamento_mm, area_mm2
```

```python
def classificar_movimento(deslocamento_mm):
    if deslocamento_mm < 2:
        return "Leve", "OK"
    elif deslocamento_mm < 5:
        return "Moderado", "Atenção"
    else:
        return "Crítico", "PARAR"
```

```python
def acao_equipamento(status):
    if status == "PARAR":
        return "Interromper aquisição / feixe imediatamente"
    elif status == "Atenção":
        return "Verificar paciente / possível correção"
    else:
        return "Operação normal"
```

```python
def preparar_visualizacoes(imagem_bgr, resultado):
    imagem_rgb = cv2.cvtColor(imagem_bgr, cv2.COLOR_BGR2RGB)

    mask_uint8 = (resultado["mask"] * 255).astype(np.uint8)

    mask_bgr = cv2.cvtColor(mask_uint8, cv2.COLOR_GRAY2BGR)
    cv2.circle(
        mask_bgr,
        (resultado["coluna_c"], resultado["linha_c"]),
        6,
        (0, 0, 255),
        -1
    )
    mask_centroide_rgb = cv2.cvtColor(mask_bgr, cv2.COLOR_BGR2RGB)

    return imagem_rgb, mask_uint8, mask_centroide_rgb
```

```python
def plotar_resultado(nome_frame, imagem_rgb, mask_uint8, mask_centroide_rgb, resultado):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    axes[0].imshow(imagem_rgb)
    axes[0].set_title(f"{nome_frame} - Imagem original")
    axes[0].axis("off")

    axes[1].imshow(mask_uint8, cmap="gray")
    axes[1].set_title(f"{nome_frame} - Máscara")
    axes[1].axis("off")

    axes[2].imshow(mask_centroide_rgb)
    axes[2].set_title(
        f"{nome_frame} - Centroide\n"
        f"L={resultado['centroide_linha']:.1f}, C={resultado['centroide_coluna']:.1f}"
    )
    axes[2].axis("off")

    plt.tight_layout()
    plt.show()
```

```python
resultados = []

referencia = None

for i, (nome_frame, arquivo) in enumerate(lista_arquivos):
    imagem_bgr = carregar_imagem(arquivo)
    resultado = detectar_paciente_mask(imagem_bgr, predictor)

    if resultado is None:
        print(f"Nenhuma pessoa detectada em {nome_frame}.")
        continue

    if i == 0:
        referencia = {
            "centroide_linha": resultado["centroide_linha"],
            "centroide_coluna": resultado["centroide_coluna"],
            "area": resultado["area"]
        }
        deslocamento = 0.0
    else:
        deslocamento = calcular_deslocamento(referencia, resultado)

    imagem_rgb, mask_uint8, mask_centroide_rgb = preparar_visualizacoes(imagem_bgr, resultado)

    plotar_resultado(nome_frame, imagem_rgb, mask_uint8, mask_centroide_rgb, resultado)

    deslocamento_mm, area_mm2 = converter_para_mm(deslocamento, resultado["area"])

    classe, status = classificar_movimento(deslocamento_mm)

    acao = acao_equipamento(status)

    resultados.append({
        "Frame": nome_frame,
        "Arquivo": arquivo,
        "Score": resultado["score"],

        "Centroide (linha)": resultado["centroide_linha"],
        "Centroide (coluna)": resultado["centroide_coluna"],

        "Área (px)": resultado["area"],
        "Área (mm²)": area_mm2,

        "Deslocamento (px)": deslocamento,
        "Deslocamento (mm)": deslocamento_mm,

        "Classificação": classe,
        "Status": status,
        "Ação do Sistema": acao
    })
```
